```python
# import dependencies
from tensorflow.keras.utils import image_dataset_from_directory
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50, ResNet101, EfficientNetB0, VGG16
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import mixed_precision
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
import gc
```

1bi. Import data and encode classes

```python
# get train and test filepaths
train_fp = '../data/seg_train'
test_fp = '../data/seg_test'

# import images with keras
train = image_dataset_from_directory(train_fp, labels='inferred', label_mode='int',
                                     image_size=(128, 128), batch_size=None)
test = image_dataset_from_directory(test_fp, labels='inferred', label_mode='int',
                                    image_size=(128, 128), batch_size=None)

# normalize color values to [0, 1] for speed
train = train.map(lambda x, y: (x / 255.0, y))
test = test.map(lambda x, y: (x / 255.0, y))

mixed_precision.set_global_policy(mixed_precision.Policy('mixed_float16'))
```

1cii. Define augmentation functions

```python
# crop an image by percent
def crop(img, percent):
  height, width = img.shape[:2]
  start_x = int(0.5 * percent * width)
  start_y = int(0.5 * percent * height)
  end_x = width - start_x
  end_y = height - start_y
  return img[start_x:end_x, start_y:end_y]


# randomly zoom image between 10-20%
def zoom(img):
  height, width = img.shape[:2]
  x_zoom = np.random.randint(10, 20)
  y_zoom = np.random.randint(10, 20)
  return cv2.resize(img, dsize=None, fx=1 + x_zoom * 0.1, fy=1 + y_zoom * 0.1)


# rotate image by degrees
def rotate(img, degrees):
  # randomly select if degrees is positive or negative
  if np.random.random() < 0.5:
    degrees = -degrees
  height, width = img.shape[:2]
  matrix = cv2.getRotationMatrix2D((width / 2, height / 2), degrees, 1)
  return cv2.warpAffine(img, matrix, (width, height))


# flips image either on x, y, or y=x
def flip(img):
  return cv2.flip(img, flipCode=np.random.choice([-1, 0, 1]))


# adjust contrast by percent
def contrast(img, percent):
  # randomly select if contrast is positive or negative
  if np.random.random() < 0.5:
    percent = -percent
  return cv2.convertScaleAbs(img, alpha=percent)


# translate image by percent
def translate(img, percent):
  height, width = img.shape[:2]
  x_trans = int(percent * width)
  y_trans = int(percent * height)
  # randomly select if x translation is positive or negative
  if np.random.random() < 0.5:
    x_trans = -x_trans
  # randomly select if y translation is positive or negative
  if np.random.random() < 0.5:
    y_trans = -y_trans
  return cv2.warpAffine(img, np.float32([[1, 0, x_trans], [0, 1, y_trans]]), (width, height))


# resize images
def resize(img, size):
  return cv2.resize(img, size)
```

1cii. regularize the training images by randomly selecting functions

```python
# iterate through the training images
train_imgs = []
train_labels = []
for img, label in train:
  selection = np.random.randint(6)
  img = img.numpy()
  # crop by 10%
  if selection == 0:
    img = crop(img, 0.1)
    img = resize(img, (128, 128))
  # randomly zoom
  elif selection == 1:
    img = zoom(img)
    img = resize(img, (128, 128))
  # rotate by 10 degrees
  elif selection == 2:
    img = rotate(img, 10)
  # flip
  elif selection == 3:
    img = flip(img)
  # contrast by 15%
  elif selection == 4:
    img = contrast(img, 0.15)
  # translate by 10%
  else:
    img = translate(img, 0.1)
  train_imgs.append(img)
  train_labels.append(label.numpy())
train_imgs = np.array(train_imgs)
train_labels = np.array(train_labels)
```

1ciii. Set up pre-trained models

```python
# create base models
resnet50 = ResNet50(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
resnet101 = ResNet101(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
efficientnet = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))

# freeze model weights
resnet50.trainable = False
resnet101.trainable = False
efficientnet.trainable = False
vgg16.trainable = False

# modify end of models and compile
resnet50_output = Flatten()(resnet50.output)
resnet50_output = Dense(64, activation='relu')(resnet50_output)
resnet50_output = BatchNormalization()(resnet50_output)
resnet50_output = Dropout(0.2)(resnet50_output)
resnet50_output = Dense(6, activation='softmax')(resnet50_output)
resnet50_model = Model(inputs=resnet50.input, outputs=resnet50_output)
resnet50_model.compile(optimizer='adam', loss='categorical_crossentropy')
resnet50_model.summary()

resnet101_output = Flatten()(resnet101.output)
resnet101_output = Dense(64, activation='relu')(resnet101_output)
resnet101_output = BatchNormalization()(resnet101_output)
resnet101_output = Dropout(0.2)(resnet101_output)
resnet101_output = Dense(6, activation='softmax')(resnet101_output)
resnet101_model = Model(inputs=resnet101.input, outputs=resnet101_output)
resnet101_model.compile(optimizer='adam', loss='categorical_crossentropy')
resnet101_model.summary()

efficientnet_output = Flatten()(efficientnet.output)
efficientnet_output = Dense(64, activation='relu')(efficientnet_output)
efficientnet_output = BatchNormalization()(efficientnet_output)
efficientnet_output = Dropout(0.2)(efficientnet_output)
efficientnet_output = Dense(6, activation='softmax')(efficientnet_output)
efficientnet_model = Model(inputs=efficientnet.input, outputs=efficientnet_output)
efficientnet_model.compile(optimizer='adam', loss='categorical_crossentropy')
efficientnet_model.summary()

vgg16_output = Flatten()(vgg16.output)
vgg16_output = Dense(64, activation='relu')(vgg16_output)
vgg16_output = BatchNormalization()(vgg16_output)
vgg16_output = Dropout(0.2)(vgg16_output)
vgg16_output = Dense(6, activation='softmax')(vgg16_output)
vgg16_model = Model(inputs=vgg16.input, outputs=vgg16_output)
vgg16_model.compile(optimizer='adam', loss='categorical_crossentropy')
vgg16_model.summary()
```

```python
# create validation set
X_train, X_val, y_train, y_val = train_test_split(
    train_imgs, train_labels, test_size=0.2, stratify=train_labels)
y_train = to_categorical(y_train, num_classes=6)
y_val = to_categorical(y_val, num_classes=6)

# create early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', restore_best_weights=True, patience=5,
                               start_from_epoch=50)

# free up memory
del train
del train_imgs
del train_labels
del img
del label
del selection
del train_fp
del test_fp
del resnet50
del resnet101
del efficientnet
del vgg16
del resnet50_output
del resnet101_output
del efficientnet_output
del vgg16_output
gc.collect()
```

1civ. Train the networks and plot loss vs. epochs

```python
# train models and free memory
rn50_h = resnet50_model.fit(X_train, y_train, validation_data=(X_val, y_val),
    epochs=100, batch_size=32, callbacks=[early_stopping])
gc.collect()

rn101_h = resnet101_model.fit(X_train, y_train, validation_data=(X_val, y_val),
    epochs=100, batch_size=32, callbacks=[early_stopping])
gc.collect()

en_h = efficientnet_model.fit(X_train, y_train, validation_data=(X_val, y_val),
    epochs=100, batch_size=32, callbacks=[early_stopping])
gc.collect()

vgg16_h = vgg16_model.fit(X_train, y_train, validation_data=(X_val, y_val),
    epochs=100, batch_size=32, callbacks=[early_stopping])
del early_stopping
gc.collect()
```

```python
# plot training and validation losses for each model
for model in [rn50_h, rn101_h, en_h, vgg16_h]:
  train_loss = model.history['loss']
  val_loss = model.history['val_loss']
  epochs = [i for i in range(1, len(val_loss) + 1)]
  plt.figure(figsize=(8, 6))
  plt.plot(epochs, train_loss, label='Training Loss')
  plt.plot(epochs, val_loss, label='Validation Loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  model_name = ''
  if model == rn50_h:
    model_name = 'ResNet50'
  elif model == rn101_h:
    model_name = 'ResNet101'
  elif model == en_h:
    model_name = 'EfficientNetB0'
  else:
    model_name = 'VGG16'
  plt.title(f'{model_name}')
  plt.legend()
  plt.show()
```

1cv. Report training, validation, and test metrics

```python
# separate test images and labels
X_test = []
y_test = []
for img, label in test:
  X_test.append(img.numpy())
  y_test.append(label.numpy())
X_test = np.array(X_test)
y_test = np.array(y_test)
y_test = to_categorical(y_test, num_classes=6)

# get error metrics for each model on each data partition
for model in [resnet50_model, resnet101_model, efficientnet_model, vgg16_model]:
  if model == resnet50_model:
    print('ResNet50')
  elif model == resnet101_model:
    print('ResNet101')
  elif model == efficientnet_model:
    print('EfficientNetB0')
  else:
    print('VGG16')

  # get class probabilities
  y_train_pred_proba = model.predict(X_train)
  y_val_pred_proba = model.predict(X_val)
  y_test_pred_proba = model.predict(X_test)

  # get class predictions
  y_train_pred = np.argmax(y_train_pred_proba, axis=1)
  y_val_pred = np.argmax(y_val_pred_proba, axis=1)
  y_test_pred = np.argmax(y_test_pred_proba, axis=1)

  # precision, recall, F1 scores
  print('Training Set')
  print(classification_report(np.argmax(y_train, axis=1), y_train_pred, digits=3), end='\n\n')
  print('Validation Set')
  print(classification_report(np.argmax(y_val, axis=1), y_val_pred, digits=3), end='\n\n')
  print('Test Set')
  print(classification_report(np.argmax(y_test, axis=1), y_test_pred, digits=3), end='\n\n')

  # get AUCs
  train_auc = roc_auc_score(y_train, y_train_pred_proba, multi_class='ovr')
  val_auc = roc_auc_score(y_val, y_val_pred_proba, multi_class='ovr')
  test_auc = roc_auc_score(y_test, y_test_pred_proba, multi_class='ovr')
  print(f'Training AUC: {train_auc:.3f}')
  print(f'Validation AUC: {val_auc:.3f}')
  print(f'Test AUC: {test_auc:.3f}')

  del model
  print()
```

1cv. VGG16 seems to strongly outperform all other models.
